```python
import pandas as pd

excel_file = "/content/drive/MyDrive/BASE DE DATOS. Rubi Armijos.xlsx"

# Get sheet names
excel_sheets = pd.ExcelFile(excel_file).sheet_names

# Load data from the first sheet into a dataframe
df1 = pd.read_excel(excel_file, sheet_name=excel_sheets[0])
print(f"Data from sheet '{excel_sheets[0]}':")
display(df1.head())

# Load data from the second sheet into another dataframe and set the second row as header
df2 = pd.read_excel(excel_file, sheet_name=excel_sheets[1], header=1)
print(f"\nData from sheet '{excel_sheets[1]}':")
display(df2.head())
```
